# 正規分布と２項分布のヒストグラム描画

```julia
using PyPlot
```

```julia
# 正規分布関数の定義
function gauss(x, n)
    m = n / 2
    return exp.(-(x .- m).^2 ./ m) / sqrt(m * π)
end
```

```julia
N = big(1000)
M = 2^N
X = 440:560

bar(X, [binomial(N, i) / M for i in X], color="blue")
plot(X, gauss(collect(X), N), color="black", linewidth=2)
```

# 数値計算による積分の結果

```julia
using Pkg
Pkg.add("QuadGK")
using QuadGK
```

```julia
function normal(x)
    return exp(-((x - 500)^2 / 500)) / √(500 * π)
end
```

```julia
quadgk(normal, 0, 480)
```
